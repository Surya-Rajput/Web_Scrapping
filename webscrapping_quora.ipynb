```python
import json

# Let's assume 'soup' contains your BeautifulSoup object
json_data = soup.find_all("script", {"type": "application/ld+json"})

# Extract the "text" fields
texts = []

# Define a function to extract text from JSON recursively
def extract_text(obj):
    if isinstance(obj, dict):
        for key, value in obj.items():
            if key == "text":
                texts.append(value)
            else:
                extract_text(value)
    elif isinstance(obj, list):
        for item in obj:
            extract_text(item)

# Extract text from the parsed JSON data
for script in json_data:
    parsed_data = json.loads(script.string)
    extract_text(parsed_data)

# Print the extracted text
for text in texts:
    print(text)
```

```python
import json

# Let's assume 'soup' contains your BeautifulSoup object
json_data = soup.find_all("script", {"type": "application/ld+json"})

# Extract the "text" fields
questions = []
answers = []

# Define a function to extract text from JSON recursively
def extract_text(obj):
    if isinstance(obj, dict):
        for key, value in obj.items():
            if key == "text":
                if value.endswith("?"):
                    questions.append(value)  # Append the text to the questions list
                else:
                    answers.append(value)  # Append the text to the answers list
            else:
                extract_text(value)
    elif isinstance(obj, list):
        for item in obj:
            extract_text(item)

# Extract text from the parsed JSON data
for script in json_data:
    parsed_data = json.loads(script.string)
    extract_text(parsed_data)

# Print the extracted questions
print("Other Questions:")
for question in questions:
    print(question)

# Print the extracted answers
print("\nAnswers:")
for answer in answers:
    print(answer)
```
